## Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

Соотнесем таблицы между собой по столбцу customerID. Попробуем ответить на вопрос - все ли клиенты пользуются обеими услугами одновременно?

```python
#Функция покажет размеры датасетов, наличие пустых значений в поле customerID 
#и количество дублей в поле customerID
def luk(data, column):
        return data.shape, data[column].duplicated().sum(), data[column].nunique(), data[column].isna().sum()

#Списки с необходимой информацией
result_list_contract = luk(df_contract, 'customerID')
result_list_personal = luk(df_personal, 'customerID')
result_list_internet = luk(df_internet, 'customerID')
result_list_phone = luk(df_phone, 'customerID')

dict = {'Контрактов' : result_list_contract,
        'Клиентов всего:' : result_list_personal,
        'Клиентов Интернет-услуг:' : result_list_internet,
        'Клиентов услуг телефонии:' : result_list_phone}

#Соберем все в таблицу, для наглядности.
result_tab = pd.DataFrame.from_dict(dict,
                                    orient='index',
                                    columns=['shape', 'Дублей', 'Уникальных', 'Пустых'])
result_tab
```

Посмотрим на данные в общем виде, и возможные корреляции между ними.

```python
final_provider = [df_contract,  df_personal, df_internet, df_phone]

#Функция объединения датасетов
list_for_concat = []
def united(tables,column):
    for item in tables:
        if {column}.issubset(item.columns):     
            item = item.set_index(column)
            list_for_concat.append(item)
    return pd.concat(list_for_concat, axis =1)

df = united(final_provider, 'customerID')
```

## Поля дат

Приведем поля дат в подобающий вид, для этого:

Выделим категории клиентов уже отказавшихся от услуг и еще действующих.

Для действующих клиентов конечная дата не определена, поле *EndDate* содержит как даты, так и категорию No.
Изменим категорию No на дату предоставления информации 2020-02-01.

Вычислим разницу дат у полей *BeginDate* и *EndDate* для каждого клиента, и запишем в отдельное поле количество дней, которые клиент пользовался услугами.

```python
#Добавим новое поле Client, в нем категоризируем клиентов, пользуются ли они услугами или отказались.

df_client = df.copy()
df_client['Client'] = df['EndDate'].apply(lambda x: x if x == 'No' else 'Yes')

#Представим даты поля *BeginDate* как даты

df_date_changed = df_client.copy()
df_date_changed['BeginDate'] = pd.to_datetime(df_date_changed['BeginDate'],format='%Y-%m-%d')


#Функция заменяет категорию No на дату предоставления информации 2020-02-01

def change_date(data, column):
    data[column] = data[column].apply(lambda x: '2020-02-01' if x == 'No' else x)
    data[column] = pd.to_datetime(data[column],format='%Y-%m-%d')
    return data

df_date_changed2 = change_date(df_date_changed, 'EndDate')

#Добавим поле Duration, в нем укажем количество дней, которые клиенты пользовались услугами

df_date_changed2['Duration'] = (df_date_changed2['EndDate'] - df_date_changed2['BeginDate']).astype('timedelta64[D]')
df_date_changed2['Duration'] = df_date_changed2['Duration'].astype('float64')
```

## Поля с количественными признаками

Посмотрим поближе на распределение количественных признаков и на возможное наличие выбросов

Напишем функции: 
- **boxplot_values** - возвращает значения диаграммы размаха
- **plot_and_box** - строит диаграмму размаха и график плотности распределения
- **plots_and_values** - запускает вышеперечисленные функции

```python
def boxplot_values(data, column):
    upper_quartile = data[column].quantile(.75)
    lower_quartile = data[column].quantile(.25)
    iqr = upper_quartile - lower_quartile
    upper_whisker = data[column][data[column] <= upper_quartile + 1.5 * iqr].max()
    lower_whisker = data[column][data[column] >= lower_quartile - 1.5 * iqr].min()
    median = data[column].median()
    mean = np.mean(data[column])
    return lower_quartile, upper_quartile, lower_whisker, upper_whisker, median , mean

def plot_hist(data, column):

    f, (ax_box, ax_hist) = plt.subplots(2,
                                        sharex=True,
                                        gridspec_kw={'height_ratios': (.15, .85)},
                                        figsize=(8, 5))

    sns.distplot(data[column],
                 hist=True,
                 kde=True, 
                 bins=24,
                 hist_kws={'edgecolor':'black'})
    sns.boxplot(data[column],
                ax=ax_box) 

    ax_box.set(xlabel='')
    
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.title('{} distribution with hist'.format(column))
    
    plt.show()
```

```python
results1 = boxplot_values(df_prefinall[df_prefinall['Client'] == 'No'], 'MonthlyCharges')
results2 = boxplot_values(df_prefinall[df_prefinall['Client'] == 'Yes'], 'MonthlyCharges')
dict = {'Те, кто ушёл' : results1,
        'Все клиенты:' : results2}

#Соберем все в таблицу, для наглядности.
result_tab = pd.DataFrame.from_dict(dict,
                                    orient='index',
                                    columns=['Нижний квартиль', 'Верхний квартиль', 'Нижний "ус"', 'Верхний "ус"', 'Медиана', 'Среднее'])
result_tab
```

```python
#Посмотрим на распределения величин ежемесячных платежей (MonthlyCharges) тех, кто ушёл
%matplotlib inline
print('Распределения величин ежемесячных платежей (MonthlyCharges) тех, кто ушёл')
plot_hist(df_prefinall[df_prefinall['Client'] == 'No'], 'MonthlyCharges')

#Посмотрим на распределения величин ежемесячных платежей (MonthlyCharges) всех клиентов
print('Распределения величин ежемесячных платежей (MonthlyCharges) всех клиентов')
plot_hist(df_prefinall[df_prefinall['Client'] == 'Yes'], 'MonthlyCharges')
```

**Вывод:**
- Минимальные и максимальные значения ежемесячных платежей практически одинаковы для обеих групп;

- Однако 25%-е квартили показывают, что ушедшие тратили в месяц значительно меньше(почти 2 раза), чем оставшиеся.
Это хорошо видно на графике, большое число ушедших пользовались самыми дешевыми услугой/услугами и не пользовались дополнительными. Скорее всего это люди на пробном периоде, прощупывающие разных операторов;

- В тоже время 75%-е квартили показывают практически одинаковые значения, что может говорить о наличии в ушедних - перебежчиков(люди пользовавшиеся услугами какое то время, но решившие сменить оператора).

- Из характера распределения, можно предполоить что:
В группе ушедших клиентов - две подгруппы пробников и перебежчиков, в группе лояльных клиентов - 3 или 4 типа клиентов по уровню трат (потребляемых услуг);

- Средняя и медиана отличаются не сильно, в пользу лояльных клиентов, что логично, так как они потребляют больше услуг и готовы больше тратиться на сервисе.
