# Цели и описание задачи

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

## Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

## Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

## Поля дат

Приведем поля дат в подобающий вид, для этого:

Выделим категории клиентов уже отказавшихся от услуг и еще действующих.

Для действующих клиентов конечная дата не определена, поле *EndDate* содержит как даты, так и категорию No.
Изменим категорию No на дату предоставления информации 2020-02-01.

Вычислим разницу дат у полей *BeginDate* и *EndDate* для каждого клиента, и запишем в отдельное поле количество дней, которые клиент пользовался услугами.

```python
#Добавим новое поле Client, в нем категоризируем клиентов, пользуются ли они услугами или отказались.

df_client = df.copy()
df_client['Client'] = df['EndDate'].apply(lambda x: x if x == 'No' else 'Yes')

#Представим даты поля *BeginDate* как даты

df_date_changed = df_client.copy()
df_date_changed['BeginDate'] = pd.to_datetime(df_date_changed['BeginDate'],format='%Y-%m-%d')


#Функция заменяет категорию No на дату предоставления информации 2020-02-01

def change_date(data, column):
    data[column] = data[column].apply(lambda x: '2020-02-01' if x == 'No' else x)
    data[column] = pd.to_datetime(data[column],format='%Y-%m-%d')
    return data

df_date_changed2 = change_date(df_date_changed, 'EndDate')

#Добавим поле Duration, в нем укажем количество дней, которые клиенты пользовались услугами

df_date_changed2['Duration'] = (df_date_changed2['EndDate'] - df_date_changed2['BeginDate']).astype('timedelta64[D]')
df_date_changed2['Duration'] = df_date_changed2['Duration'].astype('float64')
```

```python
#Составим лист полей

nan_list = (['InternetService' , 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
             'TechSupport', 'StreamingTV', 'StreamingMovies'])


def nan_categorize(data, list):
    for item in list:
        data[item] = data[item].fillna('False')
    return data

df_no_nan = df_date_changed2.copy()

df_no_nan = nan_categorize(df_no_nan, nan_list)
```

```python
def boxplot_values(data, column):
    upper_quartile = data[column].quantile(.75)
    lower_quartile = data[column].quantile(.25)
    iqr = upper_quartile - lower_quartile
    upper_whisker = data[column][data[column] <= upper_quartile + 1.5 * iqr].max()
    lower_whisker = data[column][data[column] >= lower_quartile - 1.5 * iqr].min()
    median = data[column].median()
    print ('Поле:', column, '\n'
        'Нижний квартиль: {:.1f}'.format(lower_quartile), '\n' 
           'Верхний квартиль: {:.1f}'.format(upper_quartile), '\n'
           'Нижний "ус": {:.1f}'.format(lower_whisker), '\n'
           'Верхний "ус": {:.1f}'.format(upper_whisker), '\n'
           'Медиана: {:.1f}'.format(median)
          )
def mean_var_std(data, column):
    mean = np.mean(data[column])
    variance = np.var(data[column], ddof=1)
    standard_deviation = np.std(data[column], ddof=1)
    bottom_line = mean - standard_deviation * 2
    top_line= mean + standard_deviation * 2
    print ('Среднее: {:.1f}'.format(mean), '\n' 
           'Дисперсия: {:.1f}'.format(variance), '\n'
           'Стандартное отклонение: {:.1f}, интервал 95% [-2σ: +2σ]: {:.1f} - {:.1f}'.format(standard_deviation, bottom_line, top_line)
          )
    
def plot_hist(data, column):

    f, (ax_box, ax_hist) = plt.subplots(2,
                                        sharex=True,
                                        gridspec_kw={'height_ratios': (.15, .85)},
                                        figsize=(8, 5))

    sns.distplot(data[column],
                 hist=True,
                 kde=True, 
                 bins=24,
                 hist_kws={'edgecolor':'black'})
    sns.boxplot(data[column],
                ax=ax_box) 

    ax_box.set(xlabel='')
    
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.title('{} distribution with hist'.format(column))
    
    plt.show()
    
def plots_and_values(data, column):
    plot_hist(data, column), boxplot_values(data, column), mean_var_std(data, column)
```

Разделим датафрейм на: обучающую, валидационную и тестовую выбороки в пропорциях 3:1:1, используя функцию `train_test_split` из библиетеки `sklearn.model_selection` 

Разделим выборки на `X` -  признаки и `y` — целевой признак.

Для того, чтобы отмаштабировать признаки, воспользуемся одним из методов масштабирования — стандартизации данных.  
В sklearn есть отдельная структура для стандартизации данных — **StandardScaler**  
Преобразуем обучающую и тестовую выборки функцией `transform()`.

```python
# Подготовим выборки
# Масштабируем признаки
# Сделаем это конечно же функцией

def preparation(df_input, point):
    
    data = df_input.copy()
    
    train, valid_test = train_test_split(data, test_size=0.4, random_state=12345)
    valid, test = train_test_split(valid_test, test_size=0.5, random_state=12345)
    
    def features_target_split(data, point):
        features = data.drop([point], axis=1)
        target = data[point]
        return features, target

    train_features, train_target = features_target_split(train, point)
    valid_features, valid_target = features_target_split(valid, point)
    test_features, test_target = features_target_split(test, point)
    
    def data_to_StandardScaler(data, train_features):
        scaler = StandardScaler().fit(train_features)
        scaled_features = scaler.transform(data)
        scaled_features_df = pd.DataFrame(scaled_features,  index=data.index, columns=data.columns)
        return scaled_features_df
    
    valid_features_scale = data_to_StandardScaler(valid_features, train_features)
    train_features_scale = data_to_StandardScaler(train_features, train_features)
    test_features_scale = data_to_StandardScaler(test_features, train_features)
    
    print('Размер тренировочной выборки:', train_features_scale.shape[0])
    print('Размер валидационной выборки:', valid_features_scale.shape[0])
    print('Размер тестовой выборки:', test_features_scale.shape[0])
    
    return train_target, valid_target, test_target, valid_features_scale, train_features_scale, test_features_scale, train_features.columns


y_train, y_valid, y_test, X_valid, X_train , X_test, columns = preparation(df_ohe, 'Client')
```

Многие модели умеют работать с дисбалансом, но пойдем по другому пути. 
Чтобы сбаллансировать классы и увеличить количество данных, напишем функцию `upsample(features, target, repeat)` для формирования сблалансированной выборки:

```python
def upsample(features, target, repeat):
    
    features_zeros = pd.DataFrame(features[target == 0])
    features_ones = pd.DataFrame(features[target == 1])
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(features_upsampled,
                                                   target_upsampled,
                                                   random_state=42)
    
    return features_upsampled.values, target_upsampled
```

В качестве основной метрики мы будем рассматривать $ROC-AUC$, которая показывает вероятность того, что случайно выбранный экземпляр негативного класса будет иметь меньше вероятность быть распознанным как позитивный класс, чем случайно выбранный позитивный класс. Находится как площадь под $ROC$-кривой. 

Принимает значения от $0$ до $1$, где $1$ - идеальное предсказание алгоритма.

```python
#Функция обучения модели, подбора параметров,визуализации полученных результатов

def grid_search(model, params):
    
    #Обучение модели
    
    grid = GridSearchCV(model,
                        params,
                        cv=5,
                        scoring='roc_auc',
                        n_jobs=-1)
    
    grid.fit(X_train, y_train)
    best_params = grid.best_params_
    prediction = grid.predict(X_valid)
    
    #Classification Report
    
    precision = round(precision_score(y_valid, prediction), 3)
    recall = round(recall_score(y_valid, prediction), 3)
    roc_auc = round(roc_auc_score(y_valid, grid.predict_proba(X_valid)[:, 1]), 3)
    accuracy = round(accuracy_score(y_valid, prediction), 3)
    result_list = [precision, recall, roc_auc, accuracy]
    
    print('')
    print('Итоги обучения модели:')
    print('')
    print('- Лучшее сочетание гиперпараметров: \n  {}'.format(best_params))
    print('')
    print('- Precision модели: {}'.format(precision))
    print('')
    print('- Recall модели: {}'.format(recall))
    print('')
    print('- ROC-AUC модели: {}'.format(roc_auc))
    print('')
    print('- Accuracy модели: {}'.format(accuracy))
    print('')
    
    #ROC-AUC Curve Plot
    
    ns_probs = [0 for _ in range(len(y_valid))]
    lr_probs = grid.predict_proba(X_valid)
    lr_probs = lr_probs[:, 1]
    ns_fpr, ns_tpr, _ = roc_curve(y_valid, ns_probs)
    lr_fpr, lr_tpr, _ = roc_curve(y_valid, lr_probs)
    
    f, ax = plt.subplots(1, 2, figsize=(15, 6))
    
    ax[0].plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
    ax[0].plot(lr_fpr, lr_tpr, marker='.', label='ROC-AUC={}'.format(roc_auc))
    ax[0].set_title('Receiver Operating Characteristic')
    ax[0].set_xlabel('False Positive Rate')
    ax[0].set_ylabel('True Positive Rate')
    ax[0].legend()
    
    #Precision-Recall Curve Plot
    
    precision, recall, _ = precision_recall_curve(y_valid, prediction)
    no_skill = len(y_valid[y_valid==1]) / len(y_valid)
    
    ax[1].plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
    ax[1].plot(recall, precision, marker='.')
    ax[1].set_title('Precision-Recall curve')
    ax[1].legend(['All Positive', 'Model'])
    ax[1].set_xlabel('Recall')
    ax[1].set_ylabel('Precision')
    
    plt.show()
    
    return result_list, best_params
```

```python
def lgbm_fit(params):
    
    #Обучение модели
    
    lgbm = LGBMClassifier(n_jobs=-1, random_state=12345, class_weight='balanced', **params)
    
    lgbm.fit(X_train, y_train)
    prediction = lgbm.predict(X_test)
    
    #Classification Report
    
    precision = round(precision_score(y_test, prediction), 3)
    recall = round(recall_score(y_test, prediction), 3)
    roc_auc = round(roc_auc_score(y_test, lgbm.predict_proba(X_test)[:, 1]), 3)
    accuracy = round(accuracy_score(y_test, prediction), 3)
    
    print('')
    print('Итоги обучения модели на тестовых данных:')
    print('')
    print('- Precision модели: {}'.format(precision))
    print('')
    print('- Recall модели: {}'.format(recall))
    print('')
    print('- ROC-AUC модели: {}'.format(roc_auc))
    print('')
    print('- Accuracy модели: {}'.format(accuracy))
    print('')
    
    #ROC-AUC Curve Plot
    ns_probs = [0 for _ in range(len(y_test))]
    lr_probs = lgbm.predict_proba(X_test)
    lr_probs = lr_probs[:, 1]
    ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_probs)
    lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
    
    f, ax = plt.subplots(1, 2, figsize=(15, 6))
    
    ax[0].plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
    ax[0].plot(lr_fpr, lr_tpr, marker='.', label='ROC-AUC={}'.format(roc_auc))
    ax[0].set_title('Receiver Operating Characteristic')
    ax[0].set_xlabel('False Positive Rate')
    ax[0].set_ylabel('True Positive Rate')
    ax[0].legend()
    
    #Precision-Recall Curve Plot
    precision, recall, _ = precision_recall_curve(y_test, prediction)
    no_skill = len(y_test[y_test==1]) / len(y_test)
    
    ax[1].plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
    ax[1].plot(recall, precision, marker='.')
    ax[1].set_title('Precision-Recall Curve')
    ax[1].legend(['All Positive', 'Model'])
    ax[1].set_xlabel('Recall')
    ax[1].set_ylabel('Precision')
    
    # 
    f = plt.subplots(figsize=(12, 10))
    
    importances = lgbm.feature_importances_
    model_importances = pd.Series(importances, index=columns).sort_values(ascending=False)

    sns.barplot(x=model_importances.values,
                y=model_importances.index,
                color='#56B4E9')
    plt.title('Feature importances')
    plt.xlabel('Coefficients')
    plt.tight_layout()
    
    plt.show()
```
